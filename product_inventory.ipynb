```python
df.replace(["invalid", "Invalid", ""], np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
#cleaning the Product_ID
df["Product_ID"]
```

```python
df[df["Product_ID"].duplicated(keep=False)]
```

```python
#cleaning product_name
df["Product_Name"]
```

```python
df.fillna({"Product_Name" : "Unknown"}, inplace=True)
```

```python
df["Product_Name"] = df["Product_Name"].str.strip().str.lower()
df["Product_Name"]
```

```python
#Cleaning Category#
df["Category"]
```

```python
df.fillna({"Category" : "Unknown"}, inplace=True)
```

```python
df["Category"] = df["Category"].str.strip().str.lower()
df["Category"]
```

```python
#Cleaning Suppliers
df["Supplier"]
```

```python
df["Quantity"]
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Quantity"]
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
df["Quantity"]
```

```python
median_price = df["Quantity"].median()
df.loc[df["Quantity"] <= 0, "Quantity"] = median_price
df.loc[df["Quantity"] > 100, "Quantity"] = median_price
df["Quantity"]
```

```python
df
```

```python
#cleaning unit_price
df["Unit_Price"]
```

```python
df["Unit_Price"] = pd.to_numeric(df["Unit_Price"], errors="coerce")
df["Unit_Price"]
```

```python
median_price2 = df["Unit_Price"].median()
df["Unit_Price"] = df["Unit_Price"].fillna(median_price2)
df["Unit_Price"]
```

```python
df.loc[df["Unit_Price"] <= 0, "Unit_Price"] = median_price2
df.loc[df["Unit_Price"] > 10000, "Unit_Price"] = median_price2
df["Unit_Price"]
```

```python
df
```

```python
#Cleaning reorder_level
df["Reorder_Level"]
```

```python
#cleaning last_update
df["Last_Update"]
```

```python
df["Last_Update"] = pd.to_datetime(df["Last_Update"], errors="coerce")
df["Last_Update"]
```

```python
#Products needing restocking
restock = df[df["Quantity"] < df["Reorder_Level"]][["Product_ID","Product_Name","Quantity","Reorder_Level"]]
restock
```
